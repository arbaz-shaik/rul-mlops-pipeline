```python
import matplotlib.pyplot as plt
import numpy as np

# Fake "training loss" and "validation loss"
epochs = np.arange(1, 11)
train_loss = [0.9, 0.7, 0.5, 0.4, 0.35, 0.32, 0.3, 0.29, 0.28, 0.27]
val_loss   = [1.0, 0.85, 0.6, 0.55, 0.53, 0.52, 0.51, 0.50, 0.51, 0.52]

plt.plot(epochs, train_loss, label="Training Loss")
plt.plot(epochs, val_loss, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.show()
```
