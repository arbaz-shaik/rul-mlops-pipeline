```python
import  csv

data_to_write = [
  ['Name', 'Age', 'Grade'],
  ['Alice', 25, 'A'],
  ['Bob', 22, 'B'],
  ['Charlie', 28, 'A+']
]
```

```python


with open("exm.csv", "r+", encoding="utf8") as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerows(data_to_write)

with open("exm.csv", "r+", encoding="utf8") as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        print(row)
```
