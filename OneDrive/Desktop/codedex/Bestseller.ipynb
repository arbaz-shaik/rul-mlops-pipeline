```python


with open('Bestseller - Sheet1.csv','r+', encoding='utf-8') as csvfile:
    csvreader = csv.reader(csvfile)
    header = next(csvreader)
    maxsales= -float('inf')
    maxrow = None
    for row in csvreader:
       sales = float(row[4])
       if sales > maxsales:
           maxsales = sales
           maxrow = row
print(maxrow)
with open('Bestseller - Sheet2.csv','w', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(maxrow)
    print("done")



```
